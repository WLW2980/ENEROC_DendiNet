```python
import torch
import numpy as np
import matplotlib.pyplot as plt
import os
import sys

# return to main dir
current_directory = os.getcwd()
main_dir = os.path.dirname(current_directory)
sys.path.append(main_dir)
from utils.loss import rmse
```

## US06

```python
US06_log = torch.load('test_log_save/US06_0.pt')
```

```python
us06_rmse_gru = rmse(US06_log['GRU'], US06_log['target'])
us06_rmse_bigru = rmse(US06_log['biGRU'], US06_log['target'])
us06_rmse_lstm = rmse(US06_log['LSTM'], US06_log['target'])
us06_rmse_bilstm = rmse(US06_log['biLSTM'], US06_log['target'])
us06_rmse_rnn = rmse(US06_log['RNN'], US06_log['target'])
us06_rmse_birnn = rmse(US06_log['biRNN'], US06_log['target'])
us06_rmse_trans = rmse(US06_log['transformer'], US06_log['target4trans'])
us06_rmse_dendinet = rmse(US06_log['DendiNet'], US06_log['target4dendinet'])
```

```python
print(f'RMSE Loss for GRU model on US06: {us06_rmse_gru}')
print(f'RMSE Loss for biGRU model on US06: {us06_rmse_bigru}')
print(f'RMSE Loss for LSTM model on US06: {us06_rmse_lstm}')
print(f'RMSE Loss for biLSTM model on US06: {us06_rmse_bilstm}')
print(f'RMSE Loss for RNN model on US06: {us06_rmse_rnn}')
print(f'RMSE Loss for biRNN model on US06: {us06_rmse_birnn}')
print(f'RMSE Loss for Transformer model on US06: {us06_rmse_trans}')
print(f'RMSE Loss for DendiNet model on US06: {us06_rmse_dendinet}')
```

```python
#plt.plot(US06_log['LSTM'].cpu(),  '--', label='LSTM Predicted label')
biLSTM = plt.plot(US06_log['biLSTM'].cpu(),  '--', label='biLSTM Predicted label')
#plt.plot(US06_log['GRU'].cpu(),  '--', label='GRU Predicted label')
biGRU = plt.plot(US06_log['biGRU'].cpu(),  '--', label='biGRU Predicted label')
#plt.plot(US06_log['RNN'].cpu(),  '--', label='RNN Predicted label')
biRNN = plt.plot(US06_log['biRNN'].cpu(),  '--', label='biRNN Predicted label')

#plt.plot(US06_log['transformer'].cpu(),  '--', label='transformer Predicted label')
target = plt.plot(US06_log['target'].cpu(), color='black', linestyle='-', label='True label')
DendiNet = plt.plot(US06_log['DendiNet'].cpu(), '-.', color='red', label='DendiNet Predicted label')

plt.title('US06 Test Result')
plt.xlabel('Time (S)')
plt.ylabel('SOC (%)')
plt.legend()
```

```python
size = 2100
start_index = 5200

#plt.plot(US06_log['LSTM'].cpu(), '--', label='LSTM Predicted label')
plt.plot(US06_log['biLSTM'].cpu(), '--', label='biLSTM Predicted label')
#plt.plot(US06_log['GRU'].cpu(), '--', label='GRU Predicted label')
plt.plot(US06_log['biGRU'].cpu(), '--', label='biGRU Predicted label')
#plt.plot(US06_log['RNN'].cpu(), linestyle='solid', label='RNN Predicted label')
plt.plot(US06_log['biRNN'].cpu(), '--', label='biRNN Predicted label')

#plt.plot(US06_log['transformer'].cpu(), '--', label='transformer Predicted label')
plt.plot(US06_log['target'].cpu(), color='black', linestyle='-', linewidth=3.5, label='True label')
plt.plot(US06_log['DendiNet'].cpu(), '-.', color='red', label='DendiNet Predicted label')

plt.xlim(start_index, start_index+size)
plt.ylim(0.850, 0.912)
```
